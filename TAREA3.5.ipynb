TAREA 3.5

Suponga que usted es un ayudante de un laboratorio de física. En un experimento se pide medir la posición de una masa que cae en un plano inclinado a intervalos regulares en el tiempo. El equipo disponible puede medir las posiciones con una precisión de  4mm . Uno de los estudiantes usa los siguientes datos (en  cm ):

```python
datos = [0.9,  2. ,  3.6,  5.6,  8. , 10.9, 14.2, 18.]
```

Grafique estos datos con sus barras de error. Explique por qué el estudiante merece una nota de 1.0. Compare con los siguientes datos obtenidos por un estudiante que merece 7.0.


```python
datos_7 = [0.3,  1.7,  4.1,  5.7,  8.5, 10.7, 13.6, 17.2]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Nuestros  datos
datos = [0.9,  2. ,  3.6,  5.6,  8. , 10.9, 14.2, 18.]
datos_7 = [0.3,  1.7,  4.1,  5.7,  8.5, 10.7, 13.6, 17.2]

x = np.linspace(1.,5.,8) # tiempo
y = datos
y1 = datos_7

# Creamos nuestra figura para graficar
fig = plt.figure(figsize=(13,13)) 
fig.add_axes
ax1 = fig.add_subplot(2,2,1) # grafico 1 estudiante nota 1
ax2 = fig.add_subplot(2,2,2) # grafico 2 estudiante nota 7

# Graficamos nuestros datos
ax1.errorbar(x, y, label = 'desplazamiento', yerr = 0.4, ecolor='black', marker='o', linewidth=2, color='green', markerfacecolor='red')
ax2.errorbar(x, y1, label = 'desplazamiento', yerr = 0.4, ecolor='black', marker='s', linewidth=2, color='blue',  markerfacecolor='red')

# Añadimos lineas de tendencia a cada gráfico
ax1.legend()
ax2.legend()

# Modificamos el gráfico para que se pueda ver bien estéticamente
# Titulos para cada eje de los gráficos
ax1.set_xlabel('Tiempo(s)')
ax1.set_ylabel('Posición(cm)')
ax2.set_xlabel('Tiempo(s)')
ax2.set_ylabel('Posición(cm)')

# Titulos
fig.suptitle('Posición vs Tiempo') # Titulo de la figura
ax1.set_title('Estudiante nota 1')
ax2.set_title('Estudiante nota 7')

fig.tight_layout(pad=4.0) # separacion entre gráficos

# Mostramos el gráfico
plt.show()

print("El estudiante merece nota 1 porque se aprecia claramente que sus datos son creados, la curva es perfecta, se podria decir que no hizo el experimento")
print("también las barras de error se encuentran perfectamente centradas en la función, no deberia ocurrir en un experimento real ya que estas se desvian")
```

# Calificacion: 7.0
