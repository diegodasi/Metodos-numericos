Un cierto detector de partículas es golpeado por $100$ partículas cada minuto, todas con la misma velocidad de $0.99 c$ donde $c$ es la velocidad de la luz. El detector mide la velocidad de lás partículas con un error de $0.01 c$.

Escriba un código que simule la distribución de velocidades observadas por el detector en 10 segundos. (Pista: Consulte cómo extraer un número aleatorio de una distribución discreta).

Grafique la distribuciòn de velocidades simulada.

```python
import matplotlib.pyplot as plt
import numpy as np

lamda1 = 100/60     # Nuestro lambda que nos dice que en promedio el detector es golpeado por 100 particulas en  1 min
lamda2 = lamda1*10  # Nos dice cuantas particula en promedio golpean el detector en 10 segundos

def poisson(n, lamda):                                    # con poisson vamos a poder saber el promedio de particulas que detecta el detector en 10 segundos
    return lamda**n*np.exp(-lamda)/np.math.factorial(n)
def probs(lamda, m):
    return [poisson(n, lamda) for n in range(m+1)]

plt.figure(figsize=(6,6))  # Graficamos nuestra distribución


plt.bar(range(35+1), probs(lamda2, 35), label = 'Probabilidad de detectar n particulas en 10 segundos')   # n = 35 es suficientemente grande para nuestro--
plt.title('Distribución de Poisson para obserbar n particulas en 10 segundos')                            # --intarvalo de tiempo
plt.xlabel('Cantidad n de particulas observables en 10 segundos')
plt.ylabel('Probabilidad de observar n particuals en 10 segundos')
plt.legend()
plt.show()

```

```python
# Ahora sabemos que en 10 segundos lo más probable es que el detector detecte 16 particulas app en esos 10 segundos
# Generamos una gaussiana que tenga 100 repeticiones, que es el promedio de cuantas particulas por minuto golpean el detector
# Estas mediciones representan a la velocidad que golpea la particula contra el detector

mediciones = np.random.normal(0.99, 0.01, 100)   #la velocidad a la que choca cada partícula
```

```python
# Suponemos que de los 100 datos de la velocidad de cada particula que puede haber pegado en el detector escogeremos 16 aleatorios
# El 16 es por el máximo de particulas que puede golpear el detector en 10 segundos

medir = np.random.choice(mediciones, 16)    # aqui sacamos 16 muestras de las 100 que tenemos aleatoriamente

plt.hist(medir,16) #histograma y distribución de velocidades
plt.title('Distribución de velocidades en 10 $s$')
plt.xlabel('velocidad de las partículas en $c$')
plt.ylabel('frecuencia')
plt.show()
```
